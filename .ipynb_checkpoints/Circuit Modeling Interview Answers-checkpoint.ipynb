Q1. Calculate $I_{out_a}$ from Figure 1. </br>

Ans. Using Kirchhoff's Current Law (KCL), the sum of the currents entering a node must leave that node: </br>

$ I_{out_a} = \sum_{i=1}^{i=N}{(V_{in}(i) - 0) G(i)} = \sum_{i=1}^{i=N}{V_{in}(i) G(i)} $ </br>

No wire resistance is assumed here. </br>

Q2. Calculate $I_{out_b}$ when we are considering wire conductance between cells. </br>

Ans. Consider the voltages at the intermediate nodes to be denoted by $V_x(i)$ as shown in the following figure. </br>

![Figure2](Q2_diagram.png)

Writing the KCL equations at each of the intermediate nodes $V_x(i)$, we get: </br>

$ (V_{in}(1) - V_{x}(1))G(1) = (V_x(1) - V_x(2))G_{wire} $ </br>
$ (V_{in}(2) - V_{x}(2))G(2) + (V_x(1) - V_x(2))G_{wire} = (V_x(2) - V_x(3))G_{wire} $</br>
$ (V_{in}(3) - V_{x}(3))G(3) + (V_x(2) - V_x(3))G_{wire} = (V_x(3) - V_x(4))G_{wire} $ </br>
. </br>
. </br>
. </br>
$ (V_{in}(N-1) - V_{x}(N-1))G(N-1) + (V_x(N-2) - V_x(N-1))G_{wire} = (V_x(N-1) - V_x(N))G_{wire} $  </br>
$ (V_{in}(N) - V_{x}(N))G(N) + (V_x(N-1) - V_x(N))G_{wire} = (V_x(N) - 0)G_{wire}$  </br>
$ V_x(N)G_{wire} = I_{out_b}$ </br>

<!-- Adding all the equations above, we get: -->
<!-- $ \sum(V_{in}(i) - V_{x}(i))G(i) = V_x(N) G_{wire} = I_{out_b} $ -->
<!-- $ (V_{in}(N) - I_{out_b}/G_{wire})G(N) + (V_x(N-1) -I_{out_b}/G_{wire})G_{wire} = (I_{out_b}/G_{wire})G_{wire} $  </br> -->
<!-- $ V_x(N-1) = \frac{-V_{in}(N)}{G_{wire}} + \frac{I_{out_b}G(N)}{G_{wire}^2} + \frac{I_{out_b}}{G_{wire}}$  </br> -->
<!-- $ V_x(N-2) = -V_{in}(N-1)\frac{G(N-1)}{G_{wire}} + V_{in}(N)(\frac{G(N-1)}{G_{wire}^2} - \frac{2}{G_{wire}}) + I_{out_b}(\frac{2G(N)}{G_{wire}^2} + \frac{G(N)G(N-1)}{G_{wire}^3} + \frac{G(N-1)}{G_{wire}^2} + \frac{1}{G_{wire}})$  </br> -->

```python
from sympy import symbols, solve, init_printing
import sympy as sym

# Define symbolic variables
N = 100  # Number of nodes
Vin = sym.symbols('Vin1:%d' % (N+1))  # Creates Vin1, Vin2, Vin3, Vin4, .. VinN
Vx = sym.symbols('Vx1:%d' % (N+1))    # Creates Vx1, Vx2, Vx3, Vx4, .. VxN
Ioutb = sym.Symbol('Ioutb')
Gwire = 0.1   # 0.1S

# Assume that the conductances are sampled from a normal distribution 
# with conductance values ranging from 10uS - 100uS.
# Define the parameters for the normal distribution
mean = 55e-6  # Mean in seconds (midpoint of 10µs and 100µs)
std_dev = (100e-6 - 10e-6) / 6  # Standard deviation assuming 99.7% of values fall within the range

# Sample 100 values from the normal distribution
G = np.random.normal(mean, std_dev, 100)

# Clip the values to ensure they fall within the range of 10µs to 100µs
G_clipped = np.clip(samples, 10e-6, 100e-6)

# Define equations list
equations = []

# First node equation
equations.append((Vin[0] - Vx[0])*G_clipped[0] - (Vx[0] - Vx[1])*Gwire)

# Middle nodes equations
for i in range(1, N-1):
    equations.append(
        (Vin[i] - Vx[i])*G_clipped[i] + 
        (Vx[i-1] - Vx[i])*Gwire - 
        (Vx[i] - Vx[i+1])*Gwire
    )

# Last node equation
equations.append(
    (Vin[N-1] - Vx[N-1])*G_clipped[N-1] + 
    (Vx[N-2] - Vx[N-1])*Gwire - 
    Vx[N-1]*Gwire
)

# Output current equation
equations.append(Vx[N-1]*Gwire - Ioutb)

# Create list of variables to solve for
variables = list(Vx)
variables.append(Ioutb)

# Solve the system
solution = solve(equations, variables)

# Extract Ioutb from solution dictionary
Ioutb_expr = solution[Ioutb]
print(f"The expression for Ioutb is given by: \n\nIout_b = {Ioutb_expr} \n")

num_vectors = 50
vector_length = 100
min_value = 0  # 0 volts
max_value = 1  # 1 volt

# Generate 50 input voltage vectors, Vin(1:100)
Vin = np.random.uniform(min_value, max_value, size=(num_vectors, vector_length)) 

# Initialize array to store results
Ioutb_values = np.zeros(num_vectors)

# For each input vector
for i in range(num_vectors):
    # Create a dictionary mapping symbolic Vin variables to numerical values
    Vin_dict = {sym.symbols(f'Vin{j+1}'): Vin[i,j] for j in range(vector_length)}
    
    # Substitute the values and evaluate
    Ioutb_numerical = float(Ioutb_expr.subs(Vin_dict))
    Ioutb_values[i] = Ioutb_numerical

# Ioutb_values now contains 50 output current values

# Calculate ideal VMM I_ideal
I_ideal = np.dot(Vin, G_clipped)

# Plot the scatter plot
plt.scatter(I_ideal, Ioutb_values, color='blue', alpha=0.7)
plt.title('Scatter Plot of $I_{out_b}$ vs. $I_{ideal}$ with wire resistance')
plt.xlabel('$I_{ideal}$ (A)')
plt.ylabel('$I_{out_b}$ (A)')
plt.grid(True)
plt.show()
```

Q3. Considering two columns, as well as wire conductance between the two columns. Calculate the 2nd column current to ground $I_{out_c}$. </br>

Ans. Consider the voltages at the intermediate nodes to be denoted by $V_y(i)$ as shown in the following figure. </br>

![Figure2](Q3_diagram.png)

Writing the KCL equations at each intermediate node, $V_y(i)$, and considering the equivalent
conductance for each device as a series combination of the wire conductance, $G_{wire}$, and
the device conductance, $G(2, i)$, we get: </br>

$ (V_{in}(1) - V_{y}(1))(\frac{G_{wire}G(2, 1)}{G_{wire} + G(2, 1)}) = (V_y(1) - V_y(2))G_{wire} $ </br>
$ (V_{in}(2) - V_{y}(2))(\frac{G_{wire}G(2, 2)}{G_{wire} + G(2, 2)}) + (V_y(1) - V_y(2))G_{wire} = (V_y(2) - V_y(3))G_{wire} $</br>
$ (V_{in}(3) - V_{y}(3))(\frac{G_{wire}G(2, 3)}{G_{wire} + G(2, 3)}) + (V_y(2) - V_y(3))G_{wire} = (V_y(3) - V_y(4))G_{wire} $ </br>
. </br>
. </br>
. </br>
$ (V_{in}(N-1) - V_{y}(N-1))(\frac{G_{wire}G(2, N-1)}{G_{wire} + G(2, N-1)}) + (V_y(N-2) - V_y(N-1))G_{wire} = (V_y(N-1) - V_y(N))G_{wire} $  </br>
$ (V_{in}(N) - V_{y}(N))(\frac{G_{wire}G(2, N)}{G_{wire} + G(2, N)}) + (V_y(N-1) - V_y(N))G_{wire} = (V_y(N) - 0)G_{wire}$  </br>
$ V_y(N)G_{wire} = I_{out_c}$ </br>

where $G(2, i)$ denote the conductances in the 2nd column.

```python
# Define symbolic variables
N = 100  # Number of nodes
Vin = sym.symbols('Vin1:%d' % (N+1))  # Creates Vin1, Vin2, Vin3, Vin4, .. VinN
Vy = sym.symbols('Vy1:%d' % (N+1))    # Creates Vy1, Vy2, Vy3, Vy4, .. VyN
Ioutc = sym.Symbol('Ioutc')
Gwire = 0.1   # 0.1S

# Assume that the conductances are sampled from a normal distribution 
# with conductance values ranging from 10uS - 100uS.
# Define the parameters for the normal distribution
mean = 55e-6  # Mean in seconds (midpoint of 10µs and 100µs)
std_dev = (100e-6 - 10e-6) / 6  # Standard deviation assuming 99.7% of values fall within the range

# Sample 100 values from the normal distribution
G = np.random.normal(mean, std_dev, 100)

# Clip the values to ensure they fall within the range of 10µs to 100µs
G_clipped = np.clip(samples, 10e-6, 100e-6)

# Define equations list
equations = []

# First node equation
equations.append((Vin[0] - Vy[0])*((G_clipped[0]*Gwire)/(G_clipped[0] + Gwire)) - (Vy[0] - Vy[1])*Gwire)

# Middle nodes equations
for i in range(1, N-1):
    equations.append(
        (Vin[i] - Vy[i])*((G_clipped[i]*Gwire)/(G_clipped[i] + Gwire)) + 
        (Vy[i-1] - Vy[i])*Gwire - 
        (Vy[i] - Vy[i+1])*Gwire
    )

# Last node equation
equations.append(
    (Vin[N-1] - Vy[N-1])*((G_clipped[N-1]*Gwire)/(G_clipped[N-1] + Gwire)) + 
    (Vy[N-2] - Vy[N-1])*Gwire - 
    Vy[N-1]*Gwire
)

# Output current equation
equations.append(Vy[N-1]*Gwire - Ioutc)

# Create list of variables to solve for
variables = list(Vy)
variables.append(Ioutc)

# Solve the system
solution = solve(equations, variables)

# Extract Ioutb from solution dictionary
Ioutc_expr = solution[Ioutc]
print(f"The expression for Ioutc is given by: \n\nIout_c = {Ioutc_expr} \n")

num_vectors = 50
vector_length = 100
min_value = 0  # 0 volts
max_value = 1  # 1 volt

# Generate 50 input voltage vectors, Vin(1:100)
Vin = np.random.uniform(min_value, max_value, size=(num_vectors, vector_length)) 

# Initialize array to store results
Ioutc_values = np.zeros(num_vectors)

# For each input vector
for i in range(num_vectors):
    # Create a dictionary mapping symbolic Vin variables to numerical values
    Vin_dict = {sym.symbols(f'Vin{j+1}'): Vin[i,j] for j in range(vector_length)}
    
    # Substitute the values and evaluate
    Ioutc_numerical = float(Ioutc_expr.subs(Vin_dict))
    Ioutc_values[i] = Ioutc_numerical

# Ioutc_values now contains 50 output current values

# Calculate ideal VMM I_ideal
I_ideal = np.dot(Vin, G_clipped)

# Plot the scatter plot
plt.scatter(I_ideal, Ioutc_values, color='blue', alpha=0.7)
plt.title('Scatter Plot of $I_{out_c}$ vs. $I_{ideal}$ for two columns, with wire resistance')
plt.xlabel('$I_{ideal}$ (A)')
plt.ylabel('$I_{out_c}$ (A)')
plt.grid(True)
plt.show()
```

Both 1T1R cell type (A or B) should be the **equivalent** in terms of their current outputs $I_{out_d}$ as the transistor's ON-conductance, $G_{on}$, appears in series with the device conductance, $G(i, j)$, and the relative order of T and R should not matter. So, neither of the two configurations is better in terms of reducing the difference between $I_{out_d}$ and $I_{ideal}$.
2. The computation complexity increases when we change cell from a perfect resistor to 1T1R cell because now we have to consider the transistor's ON-conductance in series with the device conductance. Moreover, the current-voltage relationship of the transistor is **non-linear**, and depends on the operating region of the transistor (linear or saturation). Thus, this increases the computational complexity because the transistor ON-conductance is now dependent on the voltage across its drain and source terminals, $V_{ds}$, and is no longer a fixed constant value.
3. To solve for currents through both the columns, $I_{out_c}$ and $I_{out_d}$, numerically, we will follow a similar approach as previously taken. We will first derive the symbolic expressions for the two currents using a symbolic solver using the set of equations derived above for columns 1 and 2. For simplicity, we will assume a constant value of the transistor ON-conductance, $G_{on}$, assuming the transistor is operating in the linear region where it has a fixed linear conductance. Once we have the expression for the currents, $I_{out_c}$ and $I_{out_d}$, we will substitute multiple input vectors and calculate the current outputs (as done previously where we considered 50 different random input vectors) and check the linearity of the Vector Matrix Multiplication by plotting those output currents against ideal VMM results, $I_{ideal} = \sum_{i=1}^{i=N} V_{in}(i)G(i)$.
